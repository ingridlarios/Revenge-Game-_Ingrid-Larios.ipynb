```python
#
f"""
October 24, 2021

Author: ingrid_larios
"""

f"""
    DocString:
    
    Introduction:
    This game is based on the tv show Revenge. Emily Thorne seeks revenge on 
    the Grayson family because as a child they took everything away from her. 
    Her father was incriminated by the Grayson's was then send to jail and 
    pressumed dead after years of being in jail. Emily was filled with anger 
    and wanted the Grayson family to pay for everything they had done 
    to her. Now that Emily is a grown woman she has created this fake life 
    and is now part of the Grayson's inner circle.
    
    Stage 1: Emily will be attending a party.
    Stage 2: Daniel wants to break up with Emily. 
    Stage 3: Complete Emily's revenge quest. 
    
    B: Knownfor bugs and/or Errors:
    None. 
    
    
Map visualization

        Party ---> Break Up ---> Revenge ---> win
         |       |             |
                 `-------------`-----------> fail            
        
"""

#Creating variables for the characters

char_1 = "Emily"
char_2 = "Daniel"
char_3 = "Victoria"

#importing package to exit game
from sys import exit

#defining helper function for "enter"
def enter():
    input(prompt = f"""<Press enter to continue>""")

#Defining funtion for enter to next part 
def intro():
    
    print(f"""
    Welcome to the Revenge game! As of now you are helping Emily make 
    important decision in order to seek her revenge!""")
    
    print(f"""
    Are you ready?""")
    
    #specifying prompt 
    enter()
    stage1()
    
    
def stage1():
    #instructions
    
   #Creating variable for players name 
    players_name =  input (prompt = f""" Welcome, please enter your name: """)
    
    print(f"""
    As you start this vengeful quest you, {players_name}, will help Emily make
    decision in order to complete her vengeful quest! But be careful, some 
    decisions could make you and Emily fail the task and risk getting exposed!""")
    
    print(f"""
    Let's get started!
    """)
    
    enter()
    
    print(f"""
    {char_3}: {char_1}, I would like you to join us to the fundraising gala we are 
    having on Saturday. Everyone will be attending, from judges, to governors, 
    and big law enforcing people. It will be a great opportunity to introduce 
    you to them and a nice way to help us raise funds.
    """)
    
    enter()
          
    print(f"""
    Emily is now worried that those important people could recognize her and ruin her plans. 
    Do you think she should say yes or say no and make up an excuse? Yes or No?""")      
        
    att1 = 2
    
    while att1 > 0:
        answer = input(prompt = """Answer: """).lower()
        
        #conditional statements will now be used
        if "yes" in answer or "y" in answer:
            print(f"""
            Party was a success and Emily was not caught by anyone!""")
            stage2()

        elif "no" in answer or "n" in answer:
            print(f"""
            {char_3} is starting to become doubtful of Emily and if she is the 
            correct one for her son. Try again!""")
            att1 -=1
            
        else:
            print(f"""
            That doesn't seem to be an answer, try again! """)
            att1 -=1
    fail()    
            
def stage2():
    print(f"""
    Things had been running smoothly until {char_1} met {char_2} for lunch. {char_2} 
    had some news to share, he was breaking up with her because he didn't feel the 
    same lately. Emily thought this would ruin all of her plans because then she 
    would be kicked out of their inner circle. What should {char_1} do? 

    A) Fake a pregnancy.
    B) Talk things out, a second chance.
    C) Leave things as is and find another way.
    """)
    
    #conditional statements will now be used
    
    att2 = 2
    #bug
    while att2 > 0: 
        answer2 = input(prompt = """"
        Pick one: """).lower()
        
        #developing a conditional statement
        if "a" in answer2 or "fake" in answer2 or "pregnancy" in answer2:
            print(f"""
            {char_2}: How is that possible!!! We have been careful!""")
            att2 -=1

        elif "b" in answer2 or "talk" in answer2 or "second chance" in answer2:
            print(f"""
            {char_1}: {char_2} Please, let's work things out. I love you we cannot 
            just leave things like this.""")
            att2 -=1 

        elif "c" in answer2 or "leave things" in answer2:
            print(f"""
            {char_1}: I can't believe this is happening but if this is what you really 
            what you want then fine. We can each go our own way!""")
            stage3()

        else:
            print(f"""
            That doesn't seem to be an answer, try again! """)
            att2 -=1
    fail()
                        
def stage3():
    print(f"""
    Bummer, they broke up!!! But you know {char_1}, she always has an ace up her sleeve. 
    {char_1} decided to become part of {char_2}'s charity and she was so good at convincing 
    and talking to people that she was welcomed into the head table when she first
    arrived into The Hampton's.
    
    Time for some revenge!!! 
    
    You and {char_1} have gotten to know some powerful and trustworthy people that will help you
    bring the {char_3} and her husband down! Everything has been planned we are only missing 
    the place, where do you think the police should arrest {char_3} and her husband?
    
    A) Fundraiser Party.
    B) Her House. 
    C) Saturday Brunch with the family.
    """)
        
    att3 = 2
    
    while att3 > 0: 
        
        answer3 = input(prompt = """"Take the final step: """).lower()
        
        #developing a conditional statement 
        if "a" in answer3 or "fundraiser" in answer3 or "party" in answer3:
            print(f"""
            You made it!!! You have helped Emily fulfill her revenge!""")
            win()


        elif "b" in answer3 or "house" in answer3:
            print(f"""
            Try again!""")
            att3 -=1
            

        elif "c" in answer3 or "saturday" in answer3 or "brunch" in answer3:
            print(f"""
            Try again!""")
            att3 -=1
            fail()    

        else:
            print(f"""
            That doesn't seem to be an answer, try again! """)
            att3 -=1
            fail()
            

   
def fail():
    print(f"""
    You have failed! You have not helped {char_1} seek her revenge. 
    Would you like to try again?
    
    Yes or no?""")
    answer4 = input(prompt = """ """).lower()
    
    if "yes" in answer4 or "y" in answer4:
        print("""
        Good luck this time! Have fun helping Emily!""")
        intro()
        
        
    elif "no" in answer4 or "n" in answer4:
        print(f"""
        Thanks for playing, See you next time!""")
        exit()
        
    else:
        print(f"""
        That doesn't seem to be an answer, try again! """)
        


def win():
    print(f"""
    Congratulations! You and {char_1} have completed her vengeful quest!!!
    Would you like to play again?
         
    1. Replay
    2. Quit""")
    
    answer5 = input (prompt = """ Answer""").lower()
    
    if "1" in answer5 or "r" in answer5 or "replay" in answer5:
        intro()
        
    elif "2" in answer5 or "n" in answer5:
        print(f"""
        Thanks for playing!""")
        quit()    
    
    else:
        quit()

    
intro()
    
    
```
